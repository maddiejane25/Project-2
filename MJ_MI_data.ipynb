```python
#Import the CSV file
MI_df = pd.read_csv('heart_attack_prediction_dataset 2.csv')
MI_df.head()
```

```python
# Get info
MI_df.info()
```

```python
MI_df.describe()
```

```python
#Drop the Patient ID column
MI_df = MI_df.drop(columns=['Patient ID'])
MI_df.head()
```

```python
# Make the blood pressure column into two columns
MI_df[['Systolic BP', 'Diastolic BP']] = MI_df['Blood Pressure'].str.split('/', expand=True)
MI_df.head()
```

```python
# Drop the blood pressure column
MI_df = MI_df.drop(columns=['Blood Pressure'])
MI_df.head()
```

```python
# Create a loop for encoding the Sex, Diet, Country, Continent and Hemisphere columns
for column in MI_df.columns:
  if MI_df[column].dtype == 'object':
    MI_df[column] = MI_df[column].astype('category')
    MI_df[column] = MI_df[column].cat.codes
```

```python
MI_df.head()
```

```python
# Import Standard Scaler
from sklearn.preprocessing import StandardScaler
```

```python
# Use StandardScaler on the Age, Cholesterol, Systolic BP, Diastolic BP, Exercise Hours Per Week, Stress Level, Sedentary Hours Per Day, Income, BMI, Triglycerides, Physical Activity Days Per Week, and Sleep Hours Per Day columns
scaled = StandardScaler().fit_transform(MI_df[['Age', 'Cholesterol', 'Systolic BP',
                                                  'Diastolic BP', 'Exercise Hours Per Week',
                                                  'Stress Level', 'Sedentary Hours Per Day',
                                                  'Income', 'BMI', 'Triglycerides', 'Physical Activity Days Per Week',
                                                  'Sleep Hours Per Day']])
scaled
```

```python
scaled_df = pd.DataFrame(scaled, columns = ['Age', 'Cholesterol', 'Systolic BP',
                                                  'Diastolic BP', 'Exercise Hours Per Week',
                                                  'Stress Level', 'Sedentary Hours Per Day',
                                                  'Income', 'BMI', 'Triglycerides', 'Physical Activity Days Per Week', 'Sleep Hours Per Day'])
scaled_df.head()
```

```python
# Replace the original data with the scaled data
MI_df['Age'] = scaled_df['Age']
MI_df['Cholesterol'] = scaled_df['Cholesterol']
MI_df['Systolic BP'] = scaled_df['Systolic BP']
MI_df['Diastolic BP'] = scaled_df['Diastolic BP']
MI_df['Exercise Hours Per Week'] = scaled_df['Exercise Hours Per Week']
MI_df['Stress Level'] = scaled_df['Stress Level']
MI_df['Sedentary Hours Per Day'] = scaled_df['Sedentary Hours Per Day']
MI_df['Income'] = scaled_df['Income']
MI_df['BMI'] = scaled_df['BMI']
MI_df['Triglycerides'] = scaled_df['Triglycerides']
MI_df['Physical Activity Days Per Week'] = scaled_df['Physical Activity Days Per Week']
MI_df['Sleep Hours Per Day'] = scaled_df['Sleep Hours Per Day']
MI_df.head()
```
